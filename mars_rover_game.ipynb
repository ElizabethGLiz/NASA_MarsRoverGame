```python
import pygame
import random

from pygame.locals import *
```

```python


class Game:
    def __init__(self, background_image, title, width, height):
        # Initialize Pygame
        

        # Set up the game screen
        self.game_screen = pygame.display.set_mode((width, height))
        pygame.display.set_caption(title)
        
        # Load background image
        self.background = pygame.image.load(background_image).convert()
        
        # Game loop control variables
        self.clock = pygame.time.Clock()

        # Define font for rendering text
        self.font = pygame.font.Font(None, 36)

    def run_game_loop(self, fps, width, height, target_rect):
        # Define colors
        WHITE = (255, 255, 255)
        RED = (255, 0, 0)
        BLACK = (0, 0, 0)

        # Load rover image
        rover_image = pygame.image.load("scratchrover_backgrounds_mars/scratchrover_sprite_mars.png")
        rover_rect = rover_image.get_rect()
        rover_rect.center = (width // 2, height // 2)

        # Define hazards
        hazards = []
        for _ in range(5):
            hazard_rect = pygame.Rect(random.randint(0, width - 50), random.randint(0, height - 50), 50, 50)
            hazards.append(hazard_rect)

        running = True
        direction = 0  # 0 for stop, 1 for forward, -1 for backward
        rover_speed = 5
        mission_result = None

        while running:
            self.game_screen.fill(WHITE)

            # Event handling
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    running = False
                elif event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_UP:
                        direction = 1
                    elif event.key == pygame.K_DOWN:
                        direction = -1
                elif event.type == pygame.KEYUP:
                    if event.key == pygame.K_UP or event.key == pygame.K_DOWN:
                        direction = 0
                elif event.type == pygame.MOUSEMOTION:  # Ignore mouse motion events
                    continue
           
            # Update rover position based on direction
            rover_rect.y += direction * rover_speed

            # Update hazards
            for hazard in hazards:
                hazard.y += rover_speed  # Hazards move downward with the rover

                # Reset hazard position if it goes off-screen
                if hazard.top >= height:
                    hazard.y = 0
                    hazard.x = random.randint(0, width - 50)

                # Draw hazards
                pygame.draw.rect(self.game_screen, RED, hazard)

            # Draw rover
            self.game_screen.blit(rover_image, rover_rect)

            # Update display
            pygame.display.flip()

            # Control frame rate
            self.clock.tick(fps)
            # Collision detection with hazards
            for hazard in hazards:
                if rover_rect.colliderect(hazard):
                    print("mission result",mission_result)
                    # Mission failed if rover collides with any hazard
                    mission_result = "Failure"
                    running = False
                    break

            # Check if rover reaches the target destination
            # if rover_rect.colliderect(target_rect):
            #     print("mission result",mission_result)
            #     # Mission success if rover reaches the target without colliding with hazards
            #     mission_result = "Success"
            #     running = False

        text = None  # Initialize text with a default value
        if mission_result == "Success":
            text = self.font.render('Mission Successful!', True, BLACK)
        elif mission_result == "Failure":
            text = self.font.render('Mission Failed!', True, BLACK)

        if text:  # Check if text is not None before blitting
            self.game_screen.blit(text, (275, 350))
        pygame.display.update()

        # Delay before quitting
        pygame.time.wait(1000)

        # Quit Pygame
        pygame.quit()


```

```python




# Define the GameObject class
class GameObject(pygame.sprite.Sprite):
    def __init__(self, image_path, x, y, width, height):
        super().__init__()
        self.image = pygame.image.load(image_path).convert_alpha()
        self.rect = self.image.get_rect()
        self.rect.x = x
        self.rect.y = y
        self.width = width
        self.height = height

```

```python
# Initialize pygame
pygame.init()
SCREEN_TITLE = "Mars Rover Driving Game"
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600


new_game = Game('scratchrover_backgrounds_mars/Ascraeus Mons.jpg', SCREEN_TITLE, SCREEN_WIDTH, SCREEN_HEIGHT)
#Create a destination
target_rect = GameObject('scratchrover_backgrounds_mars/Gusev Crater.jpg', 375, 50, 10, 10)
new_game.run_game_loop(1,SCREEN_WIDTH,SCREEN_HEIGHT,target_rect)
```
